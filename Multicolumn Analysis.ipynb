```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('BlackFriday.csv')
df.head()
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
sns.countplot(x='Age',hue='Gender',data=df)
```

```python
sns.countplot(x=df['City_Category'])
```

```python
df.groupby('City_Category').size()
```

```python
df.groupby('City_Category').size().plot(kind='pie',autopct='%.2f')
```

```python
sns.countplot(hue='City_Category',x='Age',data=df)
```

```python
sns.countplot(hue='Gender',x='Marital_Status',data=df)
```

```python
df.groupby('City_Category').sum()['Purchase'].plot(kind='pie')
```

```python
df.groupby('City_Category')['Purchase'].sum()
```

```python
df.groupby('City_Category').mean()['Purchase'].plot(kind='pie')
```
