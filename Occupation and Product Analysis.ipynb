```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('BlackFriday.csv')
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
sns.countplot(x=df['Stay_In_Current_City_Years'])
```

```python
df.groupby('Stay_In_Current_City_Years').size().plot(kind='pie',autopct='%.2f')
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years', hue = 'Gender', data = df)
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years', hue = 'Marital_Status', data = df)
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years', hue = 'City_Category', data = df)
```

```python
sns.countplot(x = 'City_Category', hue = 'Age',data = df)
```

```python
df.groupby('Stay_In_Current_City_Years')['Purchase'].sum().plot(kind = 'bar')
```

```python
df.groupby('Stay_In_Current_City_Years')['Purchase'].mean().plot(kind = 'bar')
```

#### If the total purchases are highest for a particular group (e.g., people who have stayed 3 years), it suggests that this group, as a whole, spends more compared to others. This could be due to a larger number of individuals in this group or higher individual spending.
#### If the average purchase is highest for a different group (e.g., people who have stayed 5 years), it indicates that individuals in this group tend to spend more on average per purchase than individuals in other groups.
